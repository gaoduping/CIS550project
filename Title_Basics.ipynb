```python
import pandas as pd
prefix = '/content/drive'
from google.colab import drive
drive.mount(prefix, force_remount=True)
```

```python
file_path = '/content/drive/My Drive/title.basics.tsv'
data_title_basics=pd.read_csv(file_path,sep='\t',na_values=['\\N'])
data_title_basics.shape
```

1. Make a film of multiple genres into separate rows with one single genre.

```python
data_title_basics1 = data_title_basics
data_title_basics1['genres'] = data_title_basics1['genres'].str.split(',')
data_title_basics1 = data_title_basics1.explode('genres')
data_title_basics1
```

2. Convert the data type of startYear, endYear, runtimeMin to integer.

```python
data_title_basics1[['startYear','endYear','runtimeMinutes']] = data_title_basics1[['startYear','endYear','runtimeMinutes']].apply(pd.to_numeric, errors='coerce').astype('Int32')
```

```python
data_title_basics1.to_csv("title_basics.csv")
```
