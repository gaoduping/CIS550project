```python
import pandas as pd
prefix = '/content/drive'
from google.colab import drive
drive.mount(prefix, force_remount=True)
```

```python
file_path = '/content/drive/My Drive/name.basics.tsv'
data_name_basics=pd.read_csv(file_path,sep='\t',na_values=['\\N'])
data_name_basics.shape
```

```python
data_name_basics1 = data_name_basics
data_name_basics1['knownForTitles'] = data_name_basics1['knownForTitles'].str.split(',')
data_name_basics1 = data_name_basics1.explode('knownForTitles')
data_name_basics1
```

```python
data_name_basics1[['birthYear','deathYear']] = data_name_basics1[['birthYear','deathYear']].apply(pd.to_numeric, errors='coerce').astype('Int32')
```

```python
data_name_basics1.to_csv("name_basics.csv")
```
